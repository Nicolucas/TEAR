```python
from matplotlib import pyplot as plt
from matplotlib import cm
from mpl_toolkits import axes_grid1
cmap = cm.get_cmap("PuOr")
import pandas as pd
import numpy as np
```

```python
def PlotDomain(CoorX, CoorY, Field, FieldName,TimeTxt):
    try:
      fig = plt.figure(figsize = (6, 5), constrained_layout=True)
      gs = fig.add_gridspec(1, 1)
      ax = fig.add_subplot(gs[:, :])
    except:
      fig = plt.figure(figsize = (6, 5))
      ax = fig.add_subplot(1,1,1)
    ax.set_title("{FName}".format(FName = FieldName[:-4]))
    ax.set_xlabel("X-Coordinate [m]"), ax.set_ylabel("Y-Coordinate [m]")
    ax.set_aspect('equal', 'box')
    img = ax.pcolormesh(CoorX, CoorY, Field)

    ax.annotate(text=TimeTxt,xy=[0.8,0.1], xycoords= "axes fraction")
    cbar = fig.colorbar(img, shrink=.5)
    cbar.ax.set_ylabel(FieldName)
    
    return fig, img
def SigmoidFunction(x,amp,k):
    fx =  amp * (x - x * k) / (k - abs(x) * 2.0 * k + 1.0)
    return fx
```

```python
f = open("/home/nico/Documents/TEAR/Codes_TEAR/se2dr/se2wave/stdout20201127.txt", "r")
data = [eval(line.rstrip(',\n').replace("Dist,", "")) for line in f.readlines(7679999)]

print(len(data))

x = np.array(data).T.tolist()[0]
y = np.array(data).T.tolist()[1]
c = np.array(data).T.tolist()[2]


plt.figure(figsize=(13,10))
cntr2 = plt.tricontourf(x,y,c,levels=50,cmap=cmap)
cntr2.ax.set_aspect("equal")
plt.colorbar(cntr2)


xx = np.arange(-1.e4, 1.e4, 100)
plt.plot(xx,SigmoidFunction(xx,2,-0.0002),"k-")


lev = np.arange(-15000,15000,1000).tolist()
lev.remove(0)

plt.tricontour(x,y,c,levels=lev,colors="dimgray",linestyles="dashed")
```
