```python
import numpy as np
import matplotlib.pylab as plt
```

```python

def Weight(phi,dy=50.,p=1,x=5,phi_o=0):
    return 1-(0.5*np.tanh(x*((np.abs(phi)-phi_o)))+0.5)

def annot_max(x,y, ax=None):
    x=np.array(x)
    y=np.array(y) 
    xmax = x[np.argmax(y)]
    ymax = y.max()
    text= "x={:.3f}, y={:.3f}".format(xmax, ymax)
    if not ax:
        ax=plt.gca()
    bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.72)
    arrowprops=dict(arrowstyle="->",connectionstyle="angle,angleA=0,angleB=60")
    kw = dict(xycoords='data',textcoords="axes fraction",
              arrowprops=arrowprops, bbox=bbox_props, ha="right", va="top")
    ax.annotate(text, xy=(xmax, ymax), xytext=(0.94,0.96), **kw)
```

```python
plt.figure(figsize= [10, 4],dpi=100)

philist=np.arange(-75,75,1.).tolist()


label=r"(B) $w(\phi,\phi_o=50 \cdot 0.75,A = 3\frac{p}{\delta})$"
plt.plot(philist,[Weight(phi,p=3,x=3,phi_o=50*(0.75)) for phi in philist],label = label)

label=r"(C) $w(\phi,\phi_o=50 \cdot 0.65,A = 4\frac{p}{\delta})$"
plt.plot(philist,[Weight(phi,p=3,x=4,phi_o=50*(0.65)) for phi in philist],label = label)

label=r"(D) $w(\phi,\phi_o=50 \cdot 0.65,A = 3\frac{p}{\delta})$"
plt.plot(philist,[Weight(phi,p=3,x=3,phi_o=50*(0.65)) for phi in philist],label = label)

label=r"(E) $w(\phi,\phi_o=50 \cdot 0.70,A = 3\frac{p}{\delta})$"
plt.plot(philist,[Weight(phi,p=3,x=3,phi_o=50*(0.70)) for phi in philist],label = label)

plt.vlines([50],0,0.5)

plt.xlabel("$\phi$")
plt.title(r"$w(\phi,\phi_o,A)=0.5\ tanh(A(|\phi|-\phi_o))+0.5$")
plt.grid()
plt.legend()
plt.show()
```
