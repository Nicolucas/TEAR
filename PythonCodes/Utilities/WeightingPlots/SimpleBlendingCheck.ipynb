```python
import matplotlib.pylab as plt
import numpy as np
```

```python
def Weight(phi,A, phi_o=0.65):
    return (1-(0.5*np.tanh(A*((np.abs(phi)-phi_o)))+0.5))

def WeightAlt(phi,A, phi_o=0.65):
    return ((0.5*np.tanh(A*((np.abs(phi)-phi_o)))+0.5))
```

```python
def plotweighting(philist, A, delta, phi_o, enumeration):
    label=enumeration
    plt.plot(philist,Weight(philist, A = A, phi_o = phi_o), label = label)
def plotweighting2(philist, A, delta, phi_o, enumeration):
    label=enumeration
    plt.plot(philist,WeightAlt(philist, A = A, phi_o = phi_o), label = label)
    
delta = 100.0; p= 3
A = 5/delta
phi_o = delta

philist=np.arange(-(delta),(delta),.5).tolist()
#%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%
#plotweighting(philist, A,  delta, phi_o,"$\omega(\phi,\phi_o,A)$")
plotweighting2(philist, A,  delta, phi_o,"$\omega(\phi,\phi_o,A)$")
#%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%
plt.hlines(1,-100,100)
```

```python
def plotBlending(philist, A, delta, phi_o, enumeration):
    label=enumeration
    plt.plot(philist, -20*Weight(philist, A = A, phi_o = phi_o) + 10*WeightAlt(philist, A = A, phi_o = phi_o), label = label)

    
delta = 100.0; p= 3
A = 5/delta
phi_o = 0.65*delta

philist=np.arange(-(delta),(delta),.5).tolist()
#%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%
plotBlending(philist, A,  delta, phi_o,"$\omega(\phi,\phi_o,A)$")
#%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%
#plt.hlines(1,-100,100)
```
